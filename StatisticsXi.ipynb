<h1 style = "color:blue">Use of Median as alternative of Mean</h1>

```python
marks4 = Series([45,41,46,49,48,51,39,40,43,44,50,49])
# marks = marks.sort_values()
print(len(marks4))
```

```python
print(marks4.median())
```

```python
marks5 = Series([45,41,90,49,48,51,39,100,43,44,50,49])
print(marks5.median())
```

```python
marks4.plot(kind='bar')
plt.axhline(marks4.median(), linestyle='--', label=f"Mean = {marks4.median():.2f}")
plt.axhline(marks1.mean(), linestyle='--', label=f"Mean = {marks1.mean():.2f}")
plt.show()
```

<div style="color:green; background-color:lightblue; font-size:18px">As you can see, the value of median is least affected compared to mean even in the presence of outliers. Prominently extreme outliers can even mislead median. Hence in real world differenent programming tools are used to handle outlers and then suitable statistical methods are used to calculate the central tendency.</div>

<h1 style = "color:blue">Measure of Dispersion</h1>

<h2 style = "color:blue">For data with same unit using absolute measure that is Standard Deviation</h2>

```python

data1 = Series([45, 50, 55, 60, 60, 65, 70, 75, 80, 85])

mean_value = data1.mean()
std_value = data1.std()

print("Mean:", mean_value)
print("Standard Deviation:", std_value)

```

```python


x = np.arange(len(data1))


plt.scatter(x, data1)


plt.axhline(mean_value, linestyle='--', label=f"Mean = {mean_value:.2f}")
for i in range(len(data1)):
    plt.plot([x[i], x[i]], [data1[i], mean_value], linestyle=':')

plt.axhline(mean_value + std_value, linestyle=':', label='Mean + STD')
plt.axhline(mean_value - std_value, linestyle=':', label='Mean - STD')


plt.xlabel("Observation")
plt.ylabel("Value")
plt.title("Scatter of Data Around Mean")
plt.legend()

plt.show()
```

<h2 style = "color:blue">Interpretation</h2>

<div style="color:green; background-color:lightblue; font-size:18px">Here mean lies around middle of the data and explains the overall data. Observations in data can be above or below mean value but, if it is good mean then the values almost lies closer to mean value.<br>
Now, standard deviation tell us about the average distance of observations from the mean/central value.</div>

<div style="color:green; background-color:lightblue; font-size:18px">Here the value of mean is 65, and std is 13, it implies that most of the value in the dataset lie somewhere between (65+13) and (65-13).<br>
From the plot, since most of the observations lie between bounds of +STD and -STD, the data is less variable and more consistent.</div>

<h2 style = "color:blue">For two datasets using Relative measure that is Coefficient of variation(CV)

<div style="color:green; background-color:lightblue; font-size:18px">Scores of two golfers for 10 rounds of a playing series were as following.<br>
Which golfer can be considered as consistent player?<br>
Here a player is consistent if the his score is stable and predictable. He rarely performs extremely bad or well.</div>

```python
GolferA = Series([79,78,81,76,72,72,77,74,70,78])
GolferB = Series([85,86,82,82,79,86,80,82,76,86])

```

```python
CVA = round((GolferA.std(ddof=0)/GolferA.mean())*100,2)
CVB = round((GolferB.std(ddof=0)/GolferB.mean())*100,2)
```

```python
print(f"GolferA: {CVA}% \n GolferB: {CVB}%")
```

```python


# Data
GolferA = pd.Series([79,78,81,76,72,72,77,74,70,78])
GolferB = pd.Series([85,86,82,82,79,86,80,82,76,86])

cv_A = GolferA.std() / GolferA.mean()
cv_B = GolferB.std() / GolferB.mean()

print(f"GolferA CV: {cv_A:.3f}")
print(f"GolferB CV: {cv_B:.3f}")


plt.figure(figsize=(10,6))
plt.scatter(range(1, len(GolferA)+1), GolferA, color='blue', label=f'Golfer A (CV={cv_A:.3f})')
plt.scatter(range(1, len(GolferB)+1), GolferB, color='red', label=f'Golfer B (CV={cv_B:.3f})')


plt.axhline(GolferA.mean(), color='blue', linestyle='--', alpha=0.5)
plt.axhline(GolferB.mean(), color='red', linestyle='--', alpha=0.5)


plt.xlabel('Round Number')
plt.ylabel('Score')
plt.title('Scatter Plot of Golfer Scores with Coefficient of Variation')
plt.legend()
plt.grid(True)
plt.show()
```

<div style="color:green; background-color:lightblue; font-size:18px">From the plot the scores of golferb seems to be close to its mean value. Hence he is more consistent. Consistent means the player is regular in his scores and predictable. Predictable values are more preferred in statistical Analysis.</div>

<h1 style = "color:blue">Measure of Skewness

```python
income = [15, 18, 20, 22, 25, 28, 30, 35, 40, 120]
heights = [150, 155, 158, 160, 162, 165, 168, 170, 172, 175]
scores = [95, 92, 90, 88, 85, 84, 82, 80, 60, 50]
import matplotlib.pyplot as plt
import seaborn as sns

datasets = {
    "Positive Skew (Income)": income,
    "Negative Skew (Scores)": scores,
    "Normal Distribution (Heights)": heights
}

plt.figure(figsize=(12, 4))

for i, (title, data) in enumerate(datasets.items(), 1):
    plt.subplot(1, 3, i)
    sns.histplot(data, kde=True)
    plt.title(title)

plt.tight_layout()
plt.show()
```

<div style ="background-color:lightblue; color:green">The income shows positive skewness, which implies few of the top employees earn highest in the company and tail of the curve extend towards right. mean > Median > Mode and hence skewness is positive.<br>
The score shows negative skewness, which implies only few got less marks and most of the students got high marks in the exam and the tail of the curve extend towards left. Mean < Median < Mode and hence skewness is negative.<br>
For height, as the height of the persons are uniformaly distributed in the given dataset the curve is perfect bell shaped curve and hence skewness is zero.
